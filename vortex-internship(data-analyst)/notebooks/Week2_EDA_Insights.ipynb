```python
numeric_df = df.select_dtypes(include=['int64','float64'])

plt.figure(figsize=(10,6))
sns.heatmap(numeric_df.corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(data=df, x="Sales", y="Profit")
plt.title("Sales vs Profit Relationship")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(data=df, x="Discount", y="Profit")
plt.title("Discount vs Profit Relationship")
plt.show()
```

```python
category_profit = df.groupby("Category")["Profit"].mean()

category_profit
```

```python
category_profit.plot(kind="bar", figsize=(8,5))
plt.title("Average Profit by Category")
plt.ylabel("Average Profit")
plt.show()
```

Insight 1:

Technology category generates higher average profit compared to other categories.

Insight 2:

Higher discounts show a negative impact on profit, indicating excessive discounts reduce profitability.

Insight 3:

Sales and profit have a positive relationship, but high sales do not always guarantee high profit.
